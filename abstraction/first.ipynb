```python
abstract class Boss
{
    abstract void work();
}

class employee extends Boss
{
    void work()
    {
        System.out.println("employee");
    }
} 
employee e= new employee();
e.work()
```

```python
interface Car {
    void start();
    void brake();
    void acc();
}

class Maruti implements Car {

    public void start() {
        System.out.println("Started");
    }

    public void brake() {
        System.out.println("brake");
    }

    public void acc() {
        System.out.println("acce");
    }
}


Maruti m = new Maruti();
m.start();
```
